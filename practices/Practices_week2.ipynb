### **2. network architecture configuration**

```python
from tensorflow import keras
from tensorflow.keras import layers
model = keras.Sequential([
    layers.Dense(512, activation="relu"),
    layers.Dense(10, activation="softmax")
])
```

### **3. 컴파일링**

```python
## 옵티마이저(optimizer), 손실 함수(loss function), 평가 지표(metrics)를 설정.
## optimizer: 모델이 사용할 학습 알고리즘, 모델이 손실 함수를 최소화하도록 가중치를 조정
## loss function: 모델의 예측값과 실제값 사이의 차이를 측정
## metrics: 모델의 성능을 측정하기 위한 지표
model.compile(optimizer = "rmsprop",
              loss = "sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

### **4. 이미지 데이터 준비**

```python
train_images = train_images.reshape( (60000, 28 * 28) )
train_images = train_images.astype("float32") / 255
test_images = test_images.reshape( (10000, 28 * 28) )
test_images = test_images.astype("float32") / 255
```

### **5. 모델을 fitting**

```python
## batch: 한 번에 fit처리하는 개수를 설정하는 parameter
## epochs: 전체 데이터셋을 학습하는 횟수를 설정하는 parameter
model.fit(train_images, train_labels, epochs=5, batch_size=128)
```

### **6. fitting한 모델을 가지고 예측을 수행**

```python
test_digits = test_images[0:10] ## test_digits: test_images에서 11개 데이터 가져온거
predictions = model.predict(test_digits)  ## predictions: test_digits의 예측값을 저장
predictions[0]  ## predictions[0]: predictions의 0번째 데이터의 추측값
```

```python
predictions[0].argmax()
```

```python
predictions[0][7]
```

```python
test_labels[0]
```

### **7. 모델 평가**

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"test_acc: {test_acc}")
```
